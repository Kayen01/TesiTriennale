```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Funzione per leggere il file e creare i vettori
def leggi_file(file_path):
    # Inizializza i vettori
    rxr = []
    rxc = []
    modulo_quadro = []
    potenziale_particella = []
    potenziale_statico = []
    potenziale_dinamico = []
    i = 0

    # Leggi il file riga per riga
    with open(file_path, 'r') as file:
        for riga in file:
            # Suddividi la riga in colonne
            colonne = riga.split()
            
            # Assicurati che la riga abbia il numero corretto di colonne
            if len(colonne) == 0:
                continue  # Salta la riga se non ha il numero corretto di colonne
            
            # Estrai i valori delle colonne e convertili al tipo desiderato
            if(i%2==0):
                valore_rxr = float(colonne[0])
                valore_rxc = float(colonne[1])
                valore_modulo_quadro = float(colonne[2])
                valore_potenziale_particella = float(colonne[3])
            else:
                valore_potenziale_statico = float(colonne[0])
                valore_potenziale_dinamico = float(colonne[1])
            # Aggiungi i valori ai rispettivi vettori
            if(i%2==0):
                rxr.append(valore_rxr)
                rxc.append(valore_rxc)
                modulo_quadro.append(valore_modulo_quadro)
                potenziale_particella.append(valore_potenziale_particella)
            else:
                potenziale_statico.append(valore_potenziale_statico)
                potenziale_dinamico.append(valore_potenziale_dinamico)
            i=i+1

    # Converti i vettori in array NumPy per facilitare la manipolazione dei dati
    rxr = np.array(rxr)
    rxc = np.array(rxc)
    modulo_quadro = np.array(modulo_quadro)
    potenziale_particella = np.array(potenziale_particella)
    potenziale_statico = np.array(potenziale_statico)
    potenziale_dinamico = np.array(potenziale_dinamico)
    return rxr, rxc, modulo_quadro, potenziale_particella, potenziale_statico, potenziale_dinamico 
```

```python
file_path1 = './DatiSeparated/TDH_IX_psi2dof_separated.10001' 
file_path2 = './DatiSeparated/TDH_IX_psi2dof_separated.20000' 
fig_statico = plt.figure(figsize=(15,6))

rxr, rxc, modulo_quadro, potenziale_particella, potenziale_statico, potenziale_dinamico = leggi_file(file_path1)
#print(potenziale_statico)
fig1 = fig_statico.add_subplot(121)
fig1.plot(potenziale_statico)
rxr, rxc, modulo_quadro, potenziale_particella, potenziale_statico, potenziale_dinamico = leggi_file(file_path2)
#print(potenziale_statico)
fig2 = fig_statico.add_subplot(122)
fig2.plot(potenziale_statico)

fig_statico.show()
```
